### 1.2.2 missing numerical data
let's take care of numerical data first.

LotFrontage: this is the distance between street and the property, here NaN does not seem to have a meaning, the information may just be missing. To fill it with no uncoherent value, we can set the missing input to the median LotFrontage of their Neighborhood. 

GarageYrBlt: It could mean that there is no garage we can set this value to 0.

MasVnrArea: This is the masonry veneer area in square feet. Here NaN could mean that the information is missing or that there is no masonry veneer. As most of the houses in our data does not have masonry veneer (861 over 1460) it is reasonable to set NaN to 0.

````python
print("number of properties that have no masonry veneer: {}".format(all_data[all_data["MasVnrArea"] == 0]["MasVnrArea"].count()))
````

````python
### filling missing numerical data

all_data["GarageYrBlt"] = all_data["GarageYrBlt"].fillna(0)
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
````

### 1.2.3 missing categorical data
There are two types of categorical variables:
First there are nominal variables, those variable do not have any intrinsic ordering to its categories, for example color or gender. To convert those variables in to numerical value we can use One-Hot Encoding
Sencondly there are ordinal variables one option is to encode them like numerical variable e.g "excellent  = 4, good = 3 averge =2 fair =1, poor = 0 " but this approach requires the assumption that the numerical distance between each set of subsequent categories is equal. Here, as our ordinal varibles are appreciation, we can assume this is the case.

For ordinal variables it makes sense to set NaN value to zero as for exemple:
for Alley NA means no alley access, 
for GarageQual NA means no garage...  we do that in the next section.

There are four nominal attributs that have NaN value:
MasVnrType: Masonry veneer type, we can reasonably put NaN value to None
Electrical : we put it to None       
GarageType : NaN Mean No Garage Here, we set NaN it to None 
MiscFeature :  Idem

````python
all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")
all_data["Electrical"] = all_data["Electrical"].fillna("None")
all_data["GarageType"] = all_data["GarageType"].fillna("None")
all_data["MiscFeature"] = all_data["MiscFeature"].fillna("None")
````

## 1.3 treatment of categorial variables

First we are going to apply Box cox transformation of highly skewed features. Why does it matter? : 

Asymmetrical behavior and the heteroscedasticity of the variances of attributes may lead to a bias in the model. Indeed, In that context, a factor can have an significant effect on the average Nevertheless it is not due to a true factor effect but to a non-constant variation.

To reduce the skew of our data and fight the heteroscedasticity of the variance we are going to aply on Box Cox Transformation. Here we can aply the Box Cox Transformation because we do not have negative data.

Let's try two different neural network architechtures.
For that purpose we create 2 functions to create models for a 3 hidden layer network ($#Deep_NN_model$) and for a shallow network with a simple hidden layer ($#Shallow_NN_model$). These functions will make it easier for us to test different parameter values which will be the activation function and the number of neurons in each layer.

We try a deep model as we thought it would be able to extract better features than a shallow model so we compared both to see if it was the case in practise for house pricing.

````python
seed = 8
np.random.seed(seed)

def Deep_NN_model(n_input, n_hidden1, n_hidden2, n_hidden3, activation):
    """creates a neural network with 3 hidden layers"""
    model = Sequential()
    model.add(Dense(n_hidden1, input_dim=n_input, kernel_initializer='normal', activation=activation))
    model.add(Dense(n_hidden2, kernel_initializer='normal', activation=activation))
    model.add(Dense(n_hidden3, kernel_initializer='normal', activation=activation))
    model.add(Dense(1, kernel_initializer='normal', activation=activation))
    
    return model

def Shallow_NN_model(n_input, n_hidden, activation):
    """creates a neural network with 1 hidden layers"""
    model = Sequential()
    model.add(Dense(n_hidden, input_dim=n_input, kernel_initializer='normal', activation=activation))
    model.add(Dense(1, kernel_initializer='normal', activation=activation))
    
    return model


````

<h1>Parameter tuning  with model evaluation</h1>

Our goal here is to have the best possible prediction, and as a metric to evaluate our performances we use the mean squared logarithmic error.
In order to achieve the best performance we try for both models different parameter combinations and we chose the one that will give the lowest error.
The parameters here are the number of neurons in our hidden layers. 

We evaluate the performance of each combination with a k-fold cross validation with k=5 for more precision on the loss.

Networks with more hidden neurons should perform better as they are wider and can adapt better to our dataset, but a number of neurons that would be too large could produce overfitting so we have to try all combinations to check if overfitting doesn't happen.




<h2>Parameter Tuning for shallow architecture</h2>

We check what happens in practice for the different combinations with 2 different activation functions.

````python
X=X.as_matrix()
````

````python
from sklearn.model_selection import ParameterGrid
import time

# Define the hyperparameters
hyperparameters = {
    'n_hidden': [75, 100, 150, 200, 400], 
    'activation' : ['linear', 'relu']
    }

# Compute all combinations
parameter_grid = list(ParameterGrid(hyperparameters))

# Just a table to save the results
resultsDF = pd.DataFrame(columns=['n_hidden', 'activation', 'loss'])



kfold = KFold(n_splits=5, shuffle=True, random_state=seed)
for parameter_combination in parameter_grid: 
    
    #get the current time
    current_time = time.time()
    losses = []
    for train, test in kfold.split(X,y):
        model= Shallow_NN_model(X_train.shape[1], parameter_combination['n_hidden'], parameter_combination['activation'])
        model.compile(loss='mean_squared_logarithmic_error', optimizer=keras.optimizers.Adadelta())
        model.fit(X[train], y[train], batch_size = 10, epochs = 140,verbose=0)
        losses.append( model.evaluate(X[test],y[test]))
        #losses.append( model.evaluate(X[test],y[test],verbose = 0))

    
    parameter_combination['loss'] = np.average(losses)
    parameter_combination['compute_time'] = (time.time() - current_time)/5
    
    resultsDF = resultsDF.append(parameter_combination, ignore_index=True)
    
resultsDF.sort_values(by=['loss'], ascending=True)
````

## Parameter tuning for Lasso 

Parameters to tune :

- alpha : controls the degree of sparsity of the coefficients estimated.



````python
import time
import scipy.stats 
from sklearn.model_selection import ParameterSampler

#the number of hyperparameters configurations to try 
n_iter=10

# Define the hyperparameters randomly
hyperparameters = {
    'alpha' : scipy.stats.uniform(0.001, 0.01)}



# Compute all combinations
parameter_grid = list(ParameterSampler(hyperparameters, n_iter=n_iter))

# Just a table to save the results
resultsDF = pd.DataFrame(columns=['alpha', 'score'])


for parameter_combination in parameter_grid: 
    
    #get the current time
    current_time=time.time()
    
       
    model = linear_model.Lasso(alpha=parameter_combination['alpha'])
    
    parameter_combination['score'] = np.mean(cross_val_score(model, X, y, cv=5))
    parameter_combination['compute_time'] = (time.time() - current_time)/5
    
    resultsDF = resultsDF.append(parameter_combination, ignore_index=True)
    
resultsDF.sort_values(by=['score'], ascending=False)
````

## Parameter tuning for Ridge regression

Parameters to tune :

- alpha : Regularization strength; must be a positive float. Regularization improves the conditioning of the problem and reduces the variance of the estimates. Larger values specify stronger regularization. 



````python
import time
import scipy.stats 
from sklearn.model_selection import ParameterSampler

#the number of hyperparameters configurations to try 
n_iter=20

# Define the hyperparameters randomly
hyperparameters = {
    'alpha' : scipy.stats.uniform(0.5,15)}



# Compute all combinations
parameter_grid = list(ParameterSampler(hyperparameters, n_iter=n_iter))

# Just a table to save the results
resultsDF = pd.DataFrame(columns=['alpha', 'score'])


for parameter_combination in parameter_grid: 
    
    #get the current time
    current_time = time.time()
    
    
       
    model = linear_model.Ridge(alpha=parameter_combination['alpha'])
    
    
    parameter_combination['score'] = np.mean(cross_val_score(model, X, y, cv=5))
    parameter_combination['compute_time'] = (time.time() - current_time)/5
    
    resultsDF = resultsDF.append(parameter_combination, ignore_index=True)
    
resultsDF.sort_values(by=['score'], ascending=False)
````
